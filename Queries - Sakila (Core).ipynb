# Queries

1. What query would you run to get all the customers inside city_id = 312? 
Your query should return the customers' first name, last name, email, address, and city. 

```python
c = """SELECT i.first_name,i.last_name,i.email,a.address,e.city,e.city_id
FROM customer i 
INNER JOIN address a ON a.address_id = i.customer_id
INNER JOIN city e ON e.city_id = i.customer_id
WHERE e.city_id = '312'    
;"""


pd.read_sql(c, engine)

# Reference: https://stackoverflow.com/questions/51602966/triple-join-in-sql
```

2. What query would you run to get all comedy films? Note that the genre is called the category in this schema.  
Your query should return film title, description, release year, rating, and special features.

```python
cf = """SELECT title,description,name,release_year,rating,special_features
FROM film,category
WHERE name = 'Comedy'
;"""


pd.read_sql(cf, engine)
```

3. What query would you run to get all the films that Johnny Lollobrigida was in? Your query should return the actor's last name, film title, and release year. 

```python
jl = """SELECT first_name,last_name, title, release_year 
FROM film,actor
WHERE last_name = 'Lollobrigida'
;"""


pd.read_sql(jl, engine)
```

4. What query would you run to get the first and last names of all the actors in the movie titled "Bingo Talented"?

```python
bt = """SELECT first_name,last_name, title 
FROM film,actor
WHERE title = 'Bingo Talented'
;"""


pd.read_sql(bt, engine)
```

5. What query would you run to get the customer_id associated with all payments greater than twice the average payment amount? (HINT: use 2* in your query to get twice the amount). Your result should include the customer id and the amount.

```python
ci = """SELECT customer_id, 2*AVG(amount)
FROM payment
;"""


pd.read_sql(ci, engine)
```

6. What query would you run to list the first and last names of the 5 customers who have the highest number(count) of payments? You can title the number of payments as num_payments.

```python
# DESC - descending order and ASC - ascending order

cn = """SELECT first_name, last_name, payment.amount
FROM customer,payment
ORDER BY amount DESC
LIMIT 5
;"""


pd.read_sql(cn, engine)

# Reference: https://stackoverflow.com/questions/20008638/sql-order-by-from-highest-to-lowest-value
```
