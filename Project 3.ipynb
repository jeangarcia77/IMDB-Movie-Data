## Basics

```python
basics.replace({'\\N':np.nan}, inplace=True)
```

```python
basics.dropna(subset=['runtimeMinutes'],how='any',axis=0, inplace=True)
```

```python
basics.dropna(subset=['genres'],how='any',axis=0, inplace=True)
```

```python
basics.head(5)
```

```python
basics['titleType'].value_counts()
```

```python
# Filter for value of movie within the Basics dataset

basics_movie_filter = basics[basics.titleType=='movie']

# Reference: https://towardsdatascience.com/8-ways-to-filter-pandas-dataframes-d34ba585c1b8
```

```python
# Created list of values to be removed

basics_values = ['tvEpisode', 'short', 'video','tvMovie','tvSeries','tvSpecial','tvMiniSeries','tvShort','videoGame']

basics = basics[basics.titleType.isin(basics_values) == False]

# Reference: https://www.statology.org/pandas-drop-rows-with-value/
```

```python
# Confirm only values of movie within the titleType column

basics['titleType'].value_counts()
```

```python
# Exclude movies that are included in the documentary category.
is_documentary = basics['genres'].str.contains('documentary',case=False)
basics = basics[~is_documentary]
```

```python
# Filter out startYear from 2000-2022

basics_startYear_filter = basics[(basics['startYear'] >= '2000') & (basics['startYear'] <= '2022')]

# Reference: https://datascientyst.com/filter-by-date-pandas-dataframe/
```

```python
basics_startYear_filter
```

```python
basics.info()
```

```python
basics['startYear'] = basics['startYear'].astype(int)
```

```python
# delete all rows with column

basics_startYear_drop = basics[ (basics['startYear'] < '2000') & (basics['startYear'] > '2022') ].index
basics.drop(basics_startYear_drop , inplace=True)
```

```python
basics['startYear'].value_counts()
```

```python
# Created list of values to be removed

basics_startYear_values = ['1909', '1903', '1908','1906', '1910', '1907','1911','1912',
                           '1913','1914','1915','1916','1917','1918','1919','1920','1921','1922','1923',
                          '1924','1925','1926','1927','1928','1929','1930','1931','1932','1934','1935','1936']

basics = basics[basics.startYear.isin(basics_startYear_values) == False]

# Reference: https://www.statology.org/pandas-drop-rows-with-value/
```
