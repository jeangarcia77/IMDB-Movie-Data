# Replacing \N Placeholder Values

## Ratings

```python
ratings.replace({'\\N':np.nan}, inplace=True)
```

## Basics

```python
basics.replace({'\\N':np.nan}, inplace=True)
```

```python
basics.dropna(subset=['runtimeMinutes','genres','startYear'],how='any',axis=0, inplace=True)
```

```python
basics.head(5)
```

```python
## drop endYear
basics = basics.drop(columns=['endYear'])
basics
```

```python
basics['titleType'].value_counts()
```

```python
# Filter for value of movie within the Basics dataset

basics = basics.loc[basics.titleType=='movie']
basics

# Reference: https://towardsdatascience.com/8-ways-to-filter-pandas-dataframes-d34ba585c1b8
```

```python
# Created list of values to be removed

basics_values = ['tvEpisode', 'short', 'video','tvMovie','tvSeries','tvSpecial','tvMiniSeries','tvShort','videoGame']

basics = basics[basics.titleType.isin(basics_values) == False]
basics

# Reference: https://www.statology.org/pandas-drop-rows-with-value/
```

```python
# Confirm only values of movie within the titleType column

basics['titleType'].value_counts()
```

```python
# Exclude movies that are included in the documentary category.
is_documentary = basics['genres'].str.contains('documentary',case=False)
basics = basics[~is_documentary]
```

```python
basics['startYear'] = basics['startYear'].astype(float)
```

```python
# Filter out startYear from 2000-2022

## keep startYear 2000-2022
basics = basics[(basics['startYear']>=2000)&(basics['startYear']<2023)]
basics

# Reference: https://datascientyst.com/filter-by-date-pandas-dataframe/
```

```python
basics['startYear'].value_counts()
```
